```python
import cv2
import os

# Load the video you uploaded earlier
uploaded_video = list(uploaded.keys())[0]  # Get the uploaded video filename
cap = cv2.VideoCapture(uploaded_video)

frame_count = 0
save_dir = "video_frames"
os.makedirs(save_dir, exist_ok=True)

# Extract frames from the video
while True:
    success, frame = cap.read()
    if not success:
        break
    # Save each frame as an image
    frame_name = os.path.join(save_dir, f"frame_{frame_count}.jpg")
    cv2.imwrite(frame_name, frame)
    frame_count += 1

cap.release()
print(f"Total frames extracted: {frame_count}")
```

```python
import shutil

# Zip the folder containing the frames
shutil.make_archive("video_frames", 'zip', save_dir)
```

```python
from ultralytics import YOLO
import os
import cv2

# Load the YOLOv8 model (pre-trained)
model = YOLO('yolov8n.pt')  # You can also use 'yolov8s.pt' or 'yolov8m.pt' for larger models

# Directory where the frames are stored
frames_dir = "video_frames"

# Iterate through each frame and count objects
frame_files = os.listdir(frames_dir)

for frame_file in frame_files:
    frame_path = os.path.join(frames_dir, frame_file)

    # Load the frame image
    img = cv2.imread(frame_path)

    # Run YOLOv8 inference on the frame
    results = model(img)

    # Count the number of objects detected
    num_objects = len(results[0].boxes)

    print(f"Frame {frame_file}: {num_objects} objects detected.")
```
